```python
import pandas as pd
import pyarrow as pa

ruta_archivo = r'datasets\yelp\review-001.json'
filas_por_dataframe = 1000000
contador_chunks = 0

# Itera sobre el archivo JSON leyendo en bloques y exportando cada chunk a Parquet
for chunk in pd.read_json(ruta_archivo, lines=True, chunksize=filas_por_dataframe):
    if contador_chunks < 1000000:  
        # Exporta el chunk actual a Parquet
        ruta_parquet = f'datasets\yelp\chunk_{contador_chunks + 1}.parquet'
        table = pa.Table.from_pandas(chunk)
        pa.parquet.write_table(table, ruta_parquet)
        print(f'Se ha guardado un archivo Parquet en {ruta_parquet}')
        contador_chunks += 1
    else:
        break
```
