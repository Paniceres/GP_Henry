```python
import os
import pandas as pd

def procesar_archivos_parquet(ruta_carpeta, meta):
    resultados = []  # Lista para almacenar los resultados como diccionarios

    # Iterar sobre cada archivo en la carpeta
    for archivo in os.listdir(ruta_carpeta):
        if archivo.endswith('.parquet'):  # Verificar si es un archivo Parquet
            ruta_archivo = os.path.join(ruta_carpeta, archivo)  # Obtener la ruta completa del archivo
            try:
                # Leer el archivo Parquet como DataFrame
                estados = pd.read_parquet(ruta_archivo)

                merged_df = pd.merge(estados, meta, on='gmap_id', how='left')

                count_categories = merged_df.iloc[:, 14:].sum()

                # Almacenar el resultado específico como un diccionario
                resultado = {'Archivo': archivo, **count_categories.to_dict()}
                resultados.append(resultado)

            except Exception as e:
                print(f"Error al procesar {archivo}: {e}")
    
    return resultados

# Llamada a la función con la ruta de la carpeta 'estados' y el DataFrame 'estados'
resultados_procesados = procesar_archivos_parquet(r'datasets\Google Maps\reviews-estados\Estados', meta)

# Convertir la lista de diccionarios a un DataFrame
df = pd.DataFrame(resultados_procesados)

# Guardar el DataFrame como archivo CSV
df.to_parquet('cantidad_locales_por_estado.parquet', index=False)
```
