```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
meta_datos_google = pd.read_csv(r'../datasets/processed/google/metadatos-sitios/bussiness_google.json.gz')
meta_datos_google.head(1)
```

```python
# Cargar el archivo GeoJSON
florida = gpd.read_file(r'../datasets/raw/FL.geo.json')
```

```python
# Crear un GeoDataFrame a partir del DataFrame existente
geometry = [Point(xy) for xy in zip(meta_datos_google['longitude'], meta_datos_google['latitude'])]
meta_datos_google = gpd.GeoDataFrame(meta_datos_google, geometry=geometry)

# Realizar la uni√≥n espacial
empresas_en_florida = gpd.sjoin(meta_datos_google, florida, op='within')
```
