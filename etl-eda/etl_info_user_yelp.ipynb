# Usuarios yelp

```python
import pandas as pd
```

```python
user = pd.read_parquet(r'../extras/Datasets/Yelp/user.parquet')
```

```python
user.head()
```

```python
user.info()
```

```python
user = user.iloc[::,:11]
user.drop(columns = ['funny','cool','elite'],inplace=True)

user.drop_duplicates(inplace=True)

user.rename(columns={'yelping_since': 'creation'}, inplace=True)
user.rename(columns={'average_stars': 'stars'}, inplace=True)

columnas_acomodadas = ['user_id', 'name', 'creation', 'review_count', 'useful', 'fans', 'stars', 'friends']
user = user[columnas_acomodadas]

user
```

```python
revi_yelp = pd.read_parquet(r'..\datasets\processed\yelp\reviews_yelp.parquet.gz')
revi_yelp
```

```python
user_id = revi_yelp['user_id'].tolist()

user_id = set(user_id)

user_id
```

```python
user = user[user['user_id'].isin(user_id)]
user.reset_index(drop=True, inplace=True)
```

```python
user.head()
```

```python
user.to_parquet(r'..\datasets\processed\yelp\info_user_yelp.parquet.gz', compression='gzip')
```
