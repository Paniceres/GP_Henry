```python
import pandas as pd
import gzip
import os
```

```python
user = pd.read_parquet(r'..\datasets\raw\extras\yelp\user-002.parquet')
```

```python
user.head()
```

```python
user.info()
```

```python
user = user.iloc[::,:11]
user.drop(columns = ['funny','cool','elite'],inplace=True)

user.drop_duplicates(inplace=True)

user.rename(columns={'yelping_since': 'creation'}, inplace=True)
user.rename(columns={'average_stars': 'stars'}, inplace=True)

columnas_acomodadas = ['user_id', 'name', 'creation', 'review_count', 'useful', 'fans', 'stars', 'friends']
user = user[columnas_acomodadas]

user
```

```python
revi_yelp = pd.read_parquet(r'..\..\datasets\processed\yelp\reviews_yelp.parquet.gz')
revi_yelp
```

```python
user_id = revi_yelp['user_id'].tolist()

user_id = set(user_id)

user_id
```

```python
user = user[user['user_id'].isin(user_id)]
```

```python
user
```

```python
user.to_parquet(r'..\..\datasets\processed\yelp\info_user_yelp.parquet')

# Ruta del archivo CSV que quieres comprimir con gzip
input_file = r'..\..\datasets\processed\yelp\info_user_yelp.parquet'

# Ruta de salida para el archivo comprimido con gzip
output_file = r'..\..\datasets\processed\yelp\info_user_yelp.parquet.gz'

# Comprimir el archivo CSV usando gzip
with open(input_file, 'rb') as f_in:
    with gzip.open(output_file, 'wb') as f_out:
        f_out.writelines(f_in)

# Eliminar el archivo json original si se desea
os.remove(input_file)
```
