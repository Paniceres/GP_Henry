```python
import pandas as pd
```

```python
revi = pd.read_parquet(r'..\datasets\processed\google\reviews_google.parquet.gz')

revi['resp_date'] = revi['resp_date'].fillna(revi['time'])

revi['time'] = pd.to_datetime(revi['time'])  # Convertir a tipo de dato de fecha

revi['resp_date'] = pd.to_datetime(revi['resp_date'])

# Filtrar filas a partir de 2014
revi = revi[revi['time'].dt.year >= 2014]

revi = revi[revi['resp_date'].dt.year >= 2014]

# Mostrar el DataFrame filtrado
revi
```

```python
revi.to_parquet(r'..\datasets\processed\google\reviews_google.parquet.gz', compression='gzip')
```
