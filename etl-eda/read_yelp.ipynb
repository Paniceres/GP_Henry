```python
# ObtÃ©n la lista de archivos JSON
json_files = glob.glob("..\\Google Maps\\drive-download-20231120T150233Z-001\\metadata-sitios\\*.json")

# Crea una lista vacÃ­a para almacenar los dataframes
df_list = []

# Lee cada archivo JSON y agrÃ©galo a la lista de dataframes
for file in json_files:
    df = pd.read_json(file, lines=True)
    df_list.append(df)

# Concatena todos los dataframes en la lista
df_metadata = pd.concat(df_list, ignore_index=True)
df_metadata.head()
```

```python
df_metadata.columns = df_metadata.columns.str.strip()
df_business.columns = df_business.columns.str.strip()

# Selecciona solo las columnas que te interesan
df_metadata_subset = df_metadata[['name','category']]
df_business_subset = df_business[['name', 'business_id']]

# Cambia el nombre de las columnas para asegurarte de que son Ãºnicas
df_metadata_subset = df_metadata_subset.rename(columns={'name': 'name_metadata'})
df_business_subset = df_business_subset.rename(columns={'name': 'name_business'})

# Realiza el merge
merged_df = df_metadata_subset.merge(df_business_subset, left_on='name_metadata', right_on='name_business')
merged_df
```
