# Usuarios de yelp

```python
import pandas as pd
```

```python
user = pd.read_parquet(r'../extras/Datasets/Yelp/user.parquet')
user.head()
```

```python
user.info()
```

```python
# Dropeamos las columnas que no son necesarias
user.drop(columns = ['funny','cool','elite', 'useful'],inplace=True)

# Eliminamos duplicados
user.drop_duplicates(inplace=True)

# Renombramos columnas
user.rename(columns={'yelping_since': 'creation'}, inplace=True)
user.rename(columns={'average_stars': 'stars'}, inplace=True)

# Acomodamos las columnas
columnas_acomodadas = ['user_id', 'name', 'creation', 'review_count', 'fans', 'stars', 'friends']
user = user[columnas_acomodadas]

#Cambias tipo de dato a creation a data time
user['creation'] = pd.to_datetime(user['creation'])

user
```

```python
reviews_yelp = pd.read_parquet(r'../datasets/processed/yelp/reviews_yelp.parquet.gz')
reviews_yelp
```

```python
# Neceitamos los usuarios y las estrellas para sacer la cantidad de reviews que realizo el usuario y el promedio de estrellas de nuestro dataset.
reviews_yelp = reviews_yelp[['user_id','stars']]
```

Sacamos la cantidad de reviews que tienen por usuario

```python
review_counts = reviews_yelp[['user_id']].value_counts().reset_index()
review_counts.rename(columns={'count': 'review_count'}, inplace=True)
review_counts
```

Sacamos el promedio de estrellas por usuario

```python
promedio_stars = reviews_yelp.groupby('user_id').mean().reset_index().round(2)
promedio_stars
```

Reemplazamos los valores originales de review_count y stars, por los datos que tenemos luego del filtrado por los criterios elegidos para el proyecto.

```python
user.drop(columns=['review_count','stars'],inplace=True)

user = user.merge(review_counts, on='user_id')
user = user.merge(promedio_stars, on='user_id')
```

Friends lo generamos como cantidad ya que no nos interesa quienes son, sino la cantidad que posee. Al suponer que esos so usuarios de todo Estados Unidos no chequeamos la existencia dentro de nuestro dataset.

```python
user['friends'] = user['friends'].apply(lambda x: (x.split(',')))
```

```python
for i, parte in enumerate(user['friends']):
    if parte != ['None']:
        user['friends'].iloc[i] = len(parte)
    else:
        user['friends'].iloc[i] = 0
```

Acomodamos las columnas

```python
user = user[['user_id', 'name', 'creation', 'review_count', 'fans', 'friends', 'stars']]
user
```

#### Exportamos

```python
user.to_parquet('../datasets/processed/yelp/user_yelp.parquet.gz', compression='gzip')
```
